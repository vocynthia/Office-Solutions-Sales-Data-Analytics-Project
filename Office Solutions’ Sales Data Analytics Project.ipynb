```python
import pandas as pd
import matplotlib.pyplot as plt #simpler visualization
import seaborn as sns #visualization
xl = pd.ExcelFile("/content/TableauSalesData.xlsx") #copied path
SalesData = xl.parse("Orders") 
```

```python
List = SalesData[["Region","Segment","Sub-Category","Product Name","Profit","Sales","Quantity","Order Date"]]

```

# functions

```python
def SubCatProf():
  print("Total Profits for each Sub-Category")
  SubCatProf = List[["Sub-Category","Profit","Sales"]].groupby(by=["Sub-Category"]).sum().sort_values(by = "Profit", ascending = False)
  SubCatProf["Margin"] = SubCatProf["Profit"] / SubCatProf["Sales"]
  display(SubCatProf[["Profit"]])
```

```python
def SubCatProfMar():
  print("Subcategory Profit Margin Ranking")
  SubCatProf = List[["Sub-Category","Profit","Sales"]].groupby(by=["Sub-Category"]).sum()
  SubCatProf["Margin"] = SubCatProf["Profit"] / SubCatProf["Sales"]
  display(SubCatProf[["Profit","Margin"]].sort_values(by = "Margin", ascending = False))
```

```python
def TopCopiersProf():
  print("Top 5 Most Profitable Copiers")
  Copiers1 = List.loc[(SalesData["Sub-Category"]=="Copiers")]
  Copiers = Copiers1[["Product Name","Profit","Sales"]].groupby(by=["Product Name"]).sum().sort_values(by = "Profit", ascending = False)
  Copiers["Margins"] = Copiers["Profit"] / Copiers["Sales"]
  display(Copiers[["Profit","Margins"]].head(5))
```

```python
def TopBoughtCopiers():
  print("Top 10 Most Bought Copiers")
  Copiers1 = List.loc[(SalesData["Sub-Category"]=="Copiers")]
  Copiers = Copiers1[["Product Name","Profit","Sales","Quantity"]].groupby(by=["Product Name"]).sum().sort_values(by = "Quantity", ascending = False)
  Copiers["Margins"] = Copiers["Profit"] / Copiers["Sales"]
  display(Copiers[["Profit","Margins","Quantity"]].head(10))
```

```python
def TopPhoneProf():
  print("Top 5 Most Profitable Phones")
  Phones1 = List.loc[(SalesData["Sub-Category"]=="Phones")]
  Phones = Phones1[["Product Name","Profit","Sales"]].groupby(by=["Product Name"]).sum().sort_values(by = "Profit", ascending = False)
  Phones["Margins"] = Phones["Profit"] / Phones["Sales"]
  display(Phones[["Profit","Margins"]].head(5))
```

```python
def TopLabelsProf():
  print("Top 5 Most Profitable Labels")
  Label1 = List.loc[(SalesData["Sub-Category"]=="Labels")]
  Label = Label1[["Product Name","Profit","Sales"]].groupby(by=["Product Name"]).sum().sort_values(by = "Profit", ascending = False)
  Label["Margins"] = Label["Profit"] / Label["Sales"]
  display(Label[["Profit","Margins"]].head(5))
```
